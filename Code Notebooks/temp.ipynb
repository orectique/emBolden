```python
r = Rake()

with open('../Data/corpus.txt', 'r') as f:
    corpus = f.read()

listStrings = corpus.split('.')

with open('../Data/output1.txt', 'a') as f:
    for sentence in listStrings:
        r.extract_keywords_from_text(sentence)
        sentence = sentence.strip()
        for keyword in r.get_ranked_phrases():
            sentence = sentence.replace(keyword, keyword.upper())
        #print(sentence)
        f.write(sentence)
```
